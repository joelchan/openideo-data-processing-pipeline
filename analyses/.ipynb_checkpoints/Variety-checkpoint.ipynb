Let's check the confidence intervals for these correlations

```python
def get_lowest_n(column1, column2, df):
    if df[column1].dropna().count() < df[column2].dropna().count():
        return df['column1'].dropna().count()
    else:
        return df[column2].dropna().count()

%R library('psych') #there's a nifty fisherz function in the psych package
def get_correl_confidence(var1, var2):
    r = correlation_df.ix[var1,var2]
    n = get_lowest_n(var1,var2,data_subset)
    %Rpush r
    %Rpush n
    lower, upper = %R r.con(r, n, p=.95, twotailed=TRUE)
    print "%s vs %s\n\tr = %.3f, 95%% CI (N = %i) = [%.3f to %.3f]" %(var1, var2, r, n, lower, upper)
    
get_correl_confidence('insp_dist_z_insp_mean','insp_dist_z_insp_std')
get_correl_confidence('comments_preshortlist','insp_dist_count')
get_correl_confidence('comments_preshortlist','num_shortlisted_sources')
```

Ok cool, they are all fairly far away from zero, and fairly narrow as well. I'll file this away for future thought. 

The mean and sd one in particular is theoretically interesting, because it might suggest some mechanisms for our negative relationship between mean distance and shortlist outcome (it might be the interaction between the two that's really hurting, because there's a small positive correlation between sd distance and shortlist).

The relationships between comments and num insps and num shortlisted sources may help interpret effects of comments_preshortlist.

```python
%%R
fitLRstd = glm(shortlist ~ insp_dist_z_insp_std + comments_preshortlist + num_shortlisted_sources, data=data, family = binomial)
print(summary(fitLRstd))
print(logLik(fitLRstd))
```

This is with a substantially reduced subset of the data, i.e., those with more than 1 cited inspiration, so an even more "specialized" subsample - have to be extra careful about generalizing. BUT, within this set, we see that the bivariate *positive* correlation between spread and shortlist is borne out here. The AIC is nice and small. The deviance is not comparable with the other models because the N is different.

I am now very curious to see if there is an interaction between max and sd. Let's check it out.

```python
%%R
max.sd = insp_dist_z_insp_max*insp_dist_z_insp_std
fitLRinteraction = glm(shortlist ~ insp_dist_z_insp_max + insp_dist_z_insp_std + max.sd + comments_preshortlist + num_shortlisted_sources, data=data, family = binomial)
print(summary(fitLRinteraction))
print(logLik(fitLRinteraction))
```

Not very informative. If anything, maybe it suggests that the effect of sd is where the real action is at. What happens if we take out the interaction term?

```python
%%R
fitLmaxsd = glm(shortlist ~ insp_dist_z_insp_max + insp_dist_z_insp_std + comments_preshortlist + num_shortlisted_sources, data=data, family = binomial)
print(summary(fitLmaxsd))
print(logLik(fitLmaxsd))
```

Rather similar results. The coefficient for max actually goes *up* from the model without sd (but bear in mind this is with a reduced sample - not sure how to think about it), but wiht much larger standard errors. The main thing that's consistent across the models is that sd is robust, but max and mean aren't quite as, but always trending negative.
