```python
#read in the doc data
doclevelfilename = "/Users/joelc/Dropbox/Research/dissertation/OpenIDEO/Pipeline/Challenge_and_High-level_Data/iPython intermediate inputs and outputs/DocLevel_AfterDistance.xlsx"
doclevel_df = pd.read_excel(doclevelfilename,'sheet1')

#merge in the distance data for *SOURCES*
doc_distances = doclevel_df[['nodeID','distance_z_all','distance_z_insp']] #keep only the data we care about
doc_distances.columns = ['source_ID','source_distance_z_all','source_distance_z_insp']
paths = pd.merge(paths,doc_distances,how='left')

#now the shortlist data for *TARGETS*
doc_shortlist = doclevel_df[['nodeID','shortlist']] #keep only the data we care about
doc_shortlist.columns = ['target_ID','target_shortlist']
paths = pd.merge(paths,doc_shortlist,how='left')
paths
```

```python
# create source type flag
paths['sourcetype'] = [s.split('_')[1][:1] for s in paths.source_ID]
paths.head(20)
```

```python
edges = []
for rowindex, row in paths.iterrows():
    edge = (row['source_ID'], row['target_ID'])
    if edge not in edges: edges.append(edge)
edges
```

```python
inspirations = set()
inspirations.update(paths[paths.sourcetype == 'I'].source_ID)
inspirations = sorted(inspirations)
inspirations
```

```python
def dump(listfrom, listto):
	del listto[:]
	for l in listfrom:
		listto.append(l)
	del listfrom[:]

def get_shortlist_dict(insp,edges,doc_df):
    currentstack = [insp]
    nextstack = []
    history = []
    shortlist_dict = {'source_ID':insp} #change this to 'source_ID' for easier merging next time
    shortlisted_targets = []
    for i in xrange(1,20):
        for source in currentstack: 
            for e in edges: #find all their targets
                if source == e[0] and not e[1] in history:
                    history.append(e[1])
                    nextstack.append(e[1])
        #print nextstack
        sumshortlist = 0
        for target in nextstack:
            if doc_shortlist[doc_shortlist['target_ID'] == target].target_shortlist.max() > 0:
                sumshortlist += 1
                shortlisted_targets.append((target,i))
        shortlist_dict[i] = sumshortlist
        dump(nextstack,currentstack)
    return shortlist_dict, shortlisted_targets

insp_dict = []
insp_shortlist_names_ALL = {}
for inspiration in inspirations:
    insp_shortlist_dict, insp_shorlist_names = get_shortlist_dict(inspiration,edges,paths)
    insp_dict.append(insp_shortlist_dict)
    insp_shortlist_names_ALL[inspiration] = insp_shorlist_names
insp_shortlist_df = pd.DataFrame(insp_dict)
print insp_shortlist_df
print insp_shortlist_names_ALL
```

```python
insp_num_immediate_targets = {}
for inspiration in inspirations:
    inspirationhistory = []
    num_immediate_targets = 0
    for edge in edges:
        if inspiration == edge[0] and edge not in inspirationhistory:
            inspirationhistory.append(edge)
            num_immediate_targets += 1
    insp_num_immediate_targets[inspiration] = num_immediate_targets
print insp_num_immediate_targets
```
