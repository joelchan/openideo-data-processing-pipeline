Let's check the confidence intervals for these correlations

```python
def get_lowest_n(column1, column2, df):
    if df[column1].dropna().count() < df[column2].dropna().count():
        return df['column1'].dropna().count()
    else:
        return df[column2].dropna().count()

%R library('psych') #there's a nifty fisherz function in the psych package
def get_correl_confidence(var1, var2):
    r = correlation_df.ix[var1,var2]
    n = get_lowest_n(var1,var2,data_subset)
    %Rpush r
    %Rpush n
    lower, upper = %R r.con(r, n, p=.95, twotailed=TRUE)
    print "%s vs %s\n\tr = %.3f, 95%% CI (N = %i) = [%.3f to %.3f]" %(var1, var2, r, n, lower, upper)
    
get_correl_confidence('insp_dist_z_insp_mean','insp_dist_z_insp_std')
get_correl_confidence('comments_preshortlist','insp_dist_count')
get_correl_confidence('comments_preshortlist','num_shortlisted_sources')
```

Ok cool, they are all fairly far away from zero, and fairly narrow as well. I'll file this away for future thought. 

The mean and sd one in particular is theoretically interesting, because it might suggest some mechanisms for our negative relationship between mean distance and shortlist outcome (it might be the interaction between the two that's really hurting, because there's a small positive correlation between sd distance and shortlist).

The relationships between comments and num insps and num shortlisted sources may help interpret effects of comments_preshortlist.
