We want:
1) Average number of words per comment
2) Total number of comment words per author

```python
import pandas as pd
```

```python
infile = "/Users/jchan/Desktop/Dropbox/Research/Dissertation/OpenIDEO/Pipeline/Challenge_and_High-level_Data/all_comments_data_html5.txt"
comments = pd.read_csv(infile,sep="\t")
comments.head()
```

```python
authors = set()
authors.update(comments.author_url)
authors = sorted(authors)
authors
```

```python
wordcounts = []
for rowindex, row in comments.iterrows():
    rowtext = row['text']
    rowtext = str(rowtext)
    rowwords = rowtext.split(' ')
    wordcounts.append(len(rowwords))
comments['wordcounts'] = wordcounts
comments.head(10)
```

```python
comments.to_csv('all_comments_data_2013-12-12.csv')
```

```python
author_comments = pd.pivot_table(comments, rows='author_url',aggfunc={'wordcounts':'sum','text':'count'})
author_comments.to_csv('all_comments_data_byauthor_2013-12-12.csv')
```
