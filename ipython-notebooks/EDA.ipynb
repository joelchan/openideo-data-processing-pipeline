```python
def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chart junk by stripping out unncessary plot borders and axis ticks
    
    THe top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn of all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

This is very interesting.. Plotting things in this way lets us see that there is a large amount of variability in the relationship across challenges!!!

- The slope is clearly **positive** in only 3 challenges (e-waste, voting, and social-impact), i.e., nearer = better
- The slope is clearly **negative** in 4 challenges (community-initiative, identify-celebrate, social-business, and youth-employment), i.e., nearer = worse
- The slope is unclear in the remaining 5 (bone-marrow, food-production-consumption, human-rights, vibrant-cities, web-entrepreneurs)
