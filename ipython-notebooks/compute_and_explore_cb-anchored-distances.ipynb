```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os, csv, collections

#tell pandas to display wide tables as pretty HTML tables
pd.set_option('display.width',500)
pd.set_option('display.max_columns',100)

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chart junk by stripping out unncessary plot borders and axis ticks
    
    THe top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn of all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

First thing we need to do is read in the doc-topic-weights data and initilize a dataframe to hold the document-level data

```python
# read in the doc-topic weights
weightsfilename = "/Users/jchan/Desktop/Dropbox/Research/Dissertation/OpenIDEO/Pipeline/Validation/FINAL_malletLDA/sorted_CF0_DF0_400_ASP_optim_composition-6.csv"
doc_topic_weights = {}
doc_names = []
with open(weightsfilename, 'rU') as csvfile:
    filereader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in filereader:
        if 'doc' not in row[0]: #skip the header
            docname = row[0].replace("tokenized_","").replace(".txt","")
            doc_topic_weights[docname] = [float(i) for i in row[1:]]
            doc_names.append(docname)
csvfile.close()
```

Create challenge data column

```python
doclevel_df = pd.DataFrame(doc_names,columns=['docname'])
doclevel_df['challenge'] = [c.split('_')[0] for c in doclevel_df.docname]
doclevel_df.head()
```

Now we'd like to iterate over each challenge, computing the cosine for each concept/inspiration vs its challenge brief. First, we groupby challenge. Next, we grab the cosine for each concept/inspiration vs its challenge brief, and also create a standardized (z-score) version of it. NOTE that here we are "reversing" the cosine by subtracting it from zero, so that smaller numbers (i.e., larger negative numbers) indicate *less* DISTANCE, and the closer we get to ZERO, the FARTHER we are from the reference point.

**NOTE:** For some reason the cosine of a doc vs itself is not 1! Not sure what the deal is here... We're flagging them as 'NaN' for now. I don't think there's any bug there though, just a bug in my understanding of how cosine works - I tried various versions of identity dot-products (i.e., a dot-product between a vector and itself) for a sample unit-vector, and none of them went to 1. HOLE IN MENTAL MODEL!

Also note that the code here is *extremely* hacky. I still don't know how to do the "apply"/"combine" steps in the split-apply-combine steps.

```python
# function for computing cosine
def cosine(name1, name2):
    weights1 = doc_topic_weights[name1]
    weights2 = doc_topic_weights[name2]
    return np.dot(weights1, weights2)

# compute reverse cosines (raw and normalized versions)
reversecosines = []
reversecosines_z = []
for name, group in doclevel_df.groupby(['challenge']):
    names = [n for n in group.docname]
    group['reversecosine'] = [0-cosine(names[0],n) for n in names] # compute the cosines
    group.loc[group.docname == names[0],'reversecosine']='NaN' # set the challenge brief value to missing
    group['reversecosine_z'] = [(z-group.reversecosine.mean())/group.reversecosine.std() for z in group.reversecosine] # create normalized versions
    # put them into lists to put into the master dataframe
    for i in group.reversecosine:
        reversecosines.append(i)
    for i in group.reversecosine_z:
        reversecosines_z.append(i)

# put the computed values into the master data frame
doclevel_df['reversecosine'] = reversecosines
doclevel_df['reversecosine_z'] = reversecosines_z
```

Ok, now it's time to compute concept-level distance measures! First, read the path file into a dataframe.

```python
paths_df = pd.read_csv("/Users/jchan/Desktop/Dropbox/Research/Dissertation/OpenIDEO/Pipeline/Paths/_CSVversions/paths_bone-marrow.csv")
paths_df.head()
```

Next, get data for challenge and sourcetype

```python
paths_df['challenge'] = [n.split('_')[0] for n in paths_df.seed_ID]
paths_df['sourcetype'] = [n.split('_')[1] for n in paths_df.source_ID]
paths_df['sourcetype'] = [n[0] for n in paths_df.sourcetype]
paths_df.head()
```

Now grab the source_ID cosines, both raw and z

```python
paths_df['source_dist_raw'] = [float(doclevel_df[doclevel_df.docname == s].reversecosine) for s in paths_df.source_ID]
paths_df['source_dist_z'] = [float(doclevel_df[doclevel_df.docname == s].reversecosine_z) for s in paths_df.source_ID]
paths_df.head()
```

Now get the canonical set of citing concepts

```python
citingconcepts = set()
citingconcepts.update(paths_df.source_ID)
citingconcepts = sorted(citingconcepts)
print citingconcepts
```

Now get all the concept-level measures. This is incomplete at the moment, need to account for Nas and stuff. But we're well on our way!

```python
conceptdata = []
for name, group in paths_df.groupby(['seed_ID']):
    # both kinds of sources
    rowdict = {'0_conceptname':name}
    rowdict['both_dist_raw_mean'] = group.source_dist_raw.mean()
    rowdict['both_dist_raw_max'] = np.nanmax(group.source_dist_raw)
    rowdict['both_dist_raw_std'] = group.source_dist_raw.std()
    rowdict['both_dist_z_mean'] = group.source_dist_z.mean()
    rowdict['both_dist_z_max'] = np.nanmax(group.source_dist_z)
    rowdict['both_dist_z_std'] = group.source_dist_z.std()
    rowdict['both_dist_count'] = group.source_dist_z.count()
    # concept sources
    if len(group[group.sourcetype == 'C'].source_dist_raw) > 0:
        rowdict['concept_dist_raw_mean'] = group[group.sourcetype == 'C'].source_dist_raw.mean()
        rowdict['concept_dist_raw_max'] = np.nanmax(group[group.sourcetype == 'C'].source_dist_raw)
        rowdict['concept_dist_raw_std'] = group[group.sourcetype == 'C'].source_dist_raw.std()
        rowdict['concept_dist_z_mean'] = group[group.sourcetype == 'C'].source_dist_z.mean()
        rowdict['concept_dist_z_max'] = np.nanmax(group[group.sourcetype == 'C'].source_dist_z)
        rowdict['concept_dist_z_std'] = group[group.sourcetype == 'C'].source_dist_z.std()
        rowdict['concept_dist_count'] = group[group.sourcetype == 'C'].source_dist_z.count()
    else:
        rowdict['concept_dist_raw_mean'] = "NaN"
        rowdict['concept_dist_raw_max'] = "NaN"
        rowdict['concept_dist_raw_std'] = "NaN"
        rowdict['concept_dist_z_mean'] = "NaN"
        rowdict['concept_dist_z_max'] = "NaN"
        rowdict['concept_dist_z_std'] = "NaN"
        rowdict['concept_dist_count'] = 0
    # inspiration sources
    if len(group[group.sourcetype == 'I'].source_dist_raw) > 0:
        rowdict['insp_dist_raw_mean'] = group[group.sourcetype == 'I'].source_dist_raw.mean()
        rowdict['insp_dist_raw_max'] = np.nanmax(group[group.sourcetype == 'I'].source_dist_raw)
        rowdict['insp_dist_raw_std'] = group[group.sourcetype == 'I'].source_dist_raw.std()
        rowdict['insp_dist_z_mean'] = group[group.sourcetype == 'I'].source_dist_z.mean()
        rowdict['insp_dist_z_max'] = np.nanmax(group[group.sourcetype == 'I'].source_dist_z)
        rowdict['insp_dist_z_std'] = group[group.sourcetype == 'I'].source_dist_z.std()
        rowdict['insp_dist_count'] = group[group.sourcetype == 'I'].source_dist_z.count()
    else:
        rowdict['insp_dist_raw_mean'] = "NaN"
        rowdict['insp_dist_raw_max'] = "NaN"
        rowdict['insp_dist_raw_std'] = "NaN"
        rowdict['insp_dist_z_mean'] = "NaN"
        rowdict['insp_dist_z_max'] = "NaN"
        rowdict['insp_dist_z_std'] = "NaN"
        rowdict['insp_dist_count'] = 0
    conceptdata.append(rowdict)
conceptdata_df = pd.DataFrame(conceptdata)
conceptdata_df.head()
```
